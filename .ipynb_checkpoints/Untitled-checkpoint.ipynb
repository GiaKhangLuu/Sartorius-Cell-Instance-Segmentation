```python
import mrcnn.model as modellib 
from mrcnn.config import Config
```

```python
class InferenceConfig(Config):
    
    NAME = 'l2reg-and-head'
    
    # Set batch size to 1
    GPU_COUNT = 1
    IMAGES_PER_GPU = 1
    
    # No. classes
    NUM_CLASSES = 4
    
    # Image dimensions
    IMAGE_RESIZE_MODE = "none"
    IMAGE_MIN_DIM = 576
    IMAGE_MAX_DIM = 704
    IMAGE_SHAPE = [576, 720, 3]
        
    # Mini mask
    USE_MINI_MASK = False

infer_config = InferenceConfig()
```

```python
# Init model
model = modellib.MaskRCNN(mode="inference", config=infer_config, model_dir='./')

# Load weight
model.load_weights('../Downloads/mask_rcnn_l2reg-and-head_0010.h5', by_name=True)

print('Init and load weight successfully')
```

```python
import imageio

image = imageio.imread('./sartorius-cell-instance-segmentation/train/0030fd0e6378.png')
```

```python
import numpy as np

# Func. to pad images and masks
def pad_image(image, constant_values):
    HEIGHT_TARGET, HEIGHT = 576, 520
    WIDTH_TARGET, WIDTH = 704, 704
    pad_h = (HEIGHT_TARGET - HEIGHT) // 2
    pad_w = (WIDTH_TARGET - WIDTH) // 2
    
    if len(image.shape) == 3:
        return np.pad(image, ((pad_h, pad_h), (pad_w, pad_w), (0, 0)), constant_values=constant_values)
    else:
        return np.pad(image, ((pad_h, pad_h), (pad_w, pad_w)), constant_values=constant_values)
    
image = pad_image(image, 128)

#image = np.dstack([image, image, image])
image.shape
```

```python
r = new_model.predict([image])[0]
```
