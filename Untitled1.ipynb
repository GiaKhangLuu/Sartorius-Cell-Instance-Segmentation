```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
import cv2
import skimage
import colorsys
import sys
```

```python
sys.path.append('./models/Mask_RCNN')

import mrcnn.model as modellib
from mrcnn.config import Config
from mrcnn.visualize import display_instances
import mrcnn.visualize as visualize

#sys.path.append('./browser/utils.py')
from browser.utils import pad_image, unpad_image
```

```python
HEIGHT, WIDTH = 520, 704
HEIGHT_TARGET, WIDTH_TARGET = 576, 704
MRCNN_WEIGHT = './models/weight/mask_rcnn_l2reg-and-head_0010.h5'
```

```python
class InferenceConfig(Config):
    NAME = 'l2reg-and-head'
    # Set batch size to 1
    GPU_COUNT = 1
    IMAGES_PER_GPU = 1
    
    # No. classes
    NUM_CLASSES = 4

    # Image dimensions
    IMAGE_RESIZE_MODE = "none"
    IMAGE_MIN_DIM = HEIGHT_TARGET
    IMAGE_MAX_DIM = WIDTH_TARGET
    IMAGE_SHAPE = [HEIGHT_TARGET, WIDTH_TARGET, 3]
    
    # Mini mask
    USE_MINI_MASK = False

infer_config = InferenceConfig()
    
mrcnn_model = modellib.MaskRCNN(mode="inference", config=infer_config, model_dir='./')
mrcnn_model.load_weights(MRCNN_WEIGHT, by_name=True)
```

```python
img = skimage.io.imread('./sartorius-cell-instance-segmentation/train/d1d1ca1bfb8a.png')
img = pad_image(img, 128, (HEIGHT_TARGET, WIDTH_TARGET), (HEIGHT, WIDTH))
img = np.stack([img, img, img], axis=2)
```

```python
result = mrcnn_model.detect([img])[0]

mask = result['masks']
mask = unpad_image(mask, (HEIGHT_TARGET, WIDTH_TARGET), (HEIGHT, WIDTH))
img = unpad_image(img, (HEIGHT_TARGET, WIDTH_TARGET), (HEIGHT, WIDTH))
```

```python
def rle_decode_by_image_id(image_id, shape=(520, 704)):
    SHAPE=shape
    rows = sartorius_df.loc[sartorius_df['id'] == image_id]
    
    # Image shape
    mask = np.full(shape=[len(rows), np.prod(SHAPE)], fill_value=0, dtype=np.uint8)
    
    for idx, (_, row) in enumerate(rows.iterrows()):
        s = row['annotation'].split()
        starts, lengths = [np.array(x, dtype=int) for x in (s[0:][::2], s[1:][::2])]
        starts -= 1
        ends = starts + lengths
        for lo, hi in zip(starts, ends):
            mask[idx, lo:hi] = True
        
    mask = mask.reshape([len(rows), *SHAPE])
    mask = np.moveaxis(mask, 0, 2)
    
    return mask    
```

```python
mask = rle_decode_by_image_id('9b362a5983ae')
```

```python
colors = visualize.random_colors(mask.shape[-1])
for i in range(len(colors)):
    m = mask[..., i]
    color = colors[i]
    gt = visualize.apply_mask(img, m, color)
```
